```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mn = pd.read_csv('ElectricCarData_Norm.csv')
```

```python
mn.head()
```

```python
mn.columns
```

```python
mn.drop('Model',axis=1,inplace=True)
```

```python
mn.head()
```

```python
mn.info()
```

```python
mn.describe()
```

```python
mn.head()
```

```python
mn['Range']=mn['Range'].str.replace('km','')
mn['FastCharge']=mn['FastCharge'].str.replace('km/h','')
mn['TopSpeed']=mn['TopSpeed'].str.replace('km/h','')
mn['Accel']=mn['Accel'].str.replace('sec','')
mn['Efficiency']=mn['Efficiency'].str.replace('Wh/km','')
```

```python
mn.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(22,8)
sns.barplot('Brand','PriceEuro',hue='Seats',data=mn)
```

Observation:
1. Tesla motors has high price------> More than 2000000Euros
2. Honda,MG,Mini,Kia,Fiat,Volkswagan------> Less than 5000Euros 

```python
matplotlib.rcParams['figure.figsize']=(12,6)
sns.countplot(x='Seats',data=mn)
```

1. Most of the vehicles are 5 seaters.

```python
data=mn.groupby(['Brand','Range','TopSpeed','FastCharge','Efficiency']).size().reset_index()
```

```python
data
```

```python
matplotlib.rcParams['figure.figsize']=(20,8)
sns.countplot('Brand',data=data)
```

Observation:
1. Tesla vahicles are more on the road.
2. Audi,Nissan,VW,Kia are less on the road as compare with tesla.
